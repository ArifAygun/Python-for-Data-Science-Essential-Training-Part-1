# Chapter 2 - Data Preparation Basics
## Segment 4 - Concatenating and transforming data

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
DF_obj = pd.DataFrame(np.arange(36).reshape(6,6))
DF_obj
```

```python
DF_obj_2 = pd.DataFrame(np.arange(15).reshape(5,3))
DF_obj_2
```

### Concatenating data

```python
pd.concat([DF_obj, DF_obj_2])
```

```python
pd.concat([DF_obj, DF_obj_2], axis=1) # concatenate row index values
```

### Transforming data
#### Dropping data

```python
DF_obj.drop([0,2])
```

```python
DF_obj
```

```python
DF_obj.drop([2,4], axis=1)
```

### Adding data

```python
series_obj = Series(np.arange(6))
series_obj.name = 'added_variable'
series_obj
```

```python
variable_added = DataFrame.join(DF_obj, series_obj )
variable_added
```

```python
added_table = variable_added.append(variable_added, ignore_index=False)
added_table
```

```python
added_table = variable_added.append(variable_added, ignore_index=True)
added_table
```

### Sorting data

```python
DF_sorted = DF_obj.sort_values(by=(3), ascending=False)
DF_sorted
```

```python
DF_sorted = DF_obj.sort_values(by=(3))
DF_sorted
```
